- Policy.csv 에 따르면 Social Distance를 Strong과 Weak로 나눠서 보여주고 있다.



```python
weak = time.loc[91:106]
strong = time.loc[62:90]
```

```python
weak
```

```python
weak_test_total = weak['test'].sum()
weak_confirmed_total = weak['confirmed'].sum()
weak_test_total, weak_confirmed_total
```

```python
weak_ratio = weak_confirmed_total / weak_test_total
weak_ratio
```

```python
strong
```

```python
strong_test_total = strong['test'].sum()
strong_confirmed_total = strong['confirmed'].sum()
strong_test_total, strong_confirmed_total
```

```python
strong_ratio = strong_confirmed_total / strong_test_total
strong_ratio
```

- Social Distance : weak 보다 Social Distance : strong 에서 오히려 검사 받은 인원이 더 많았고, 양성ㅍ반응을 보인 사람의 비율이 더 높았다.
